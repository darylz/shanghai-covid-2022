## New Positive Cases

```python
newpos_case_file_lst = sorted(glob.glob("CasePR/0*.txt"))
```

```python

newlvyy_pat = re.compile(".*治愈出院([0-9]+)例")
newlvfc_pat1 = re.compile(".*解除医学观察无症状感染者[0-9]+例，其中本土无症状感染者([0-9]+)例.*")
newlvfc_pat2 = re.compile(".*解除医学观察本土无症状感染者([0-9]+)例.*")

newpatn_pat1 = re.compile("病例([0-9]+)，.*居住于(.*)")
newpatn_pat2 = re.compile("病例([0-9]+)[—、]病例([0-9]+)，居住于(.*)")

newasym_pat1 = re.compile("无症状感染者([0-9]+)，.*居住于(.*)")
newasym_pat2 = re.compile("无症状感染者([0-9]+)[—、]无症状感染者([0-9]+)，居住于(.*)")

df = pd.DataFrame()
records_newlvyy = []
records_newlvfc = []
    
for filename in newpos_case_file_lst:
#     print(filename)
    date_str = "2022" + os.path.splitext(filename)[0].split('/')[-1]    
    
    with open(filename) as f:
        content_raw = f.read()
    
    content_raw = content_raw.replace("区，", "区\n")
        
    lines = content_raw.split("\n")

    part_newpatn = False
    part_newasym = False

    district = None
    location = None

    location_lst_start = False
    location_valid = False
    
    records_part = []
    df_part = pd.DataFrame()

    for line in lines:
        line = line.strip('，,，：。、 ')
            
        if "本土病例情况" in line:
            part_newpatn = True
            part_newasym = False
        elif "本土无症状感染者情况" in line:
            part_newpatn = False
            part_newasym = True
        elif "境外输入病例情况" in line:
            part_newpatn = False
            part_newasym = False
        elif "境外输入性无症状感染者情况" in line:
            part_newpatn = False
            part_newasym = False

        res = newpatn_pat1.match(line)
        if res:
            district = res.group(2)
            
            record = {'Date': date_str, 'District': district, 'Qty': 1}
            records_part.append(record)
        
        res = newpatn_pat2.match(line)
        if res:
            start = int(res.group(1))
            end   = int(res.group(2))
            district = res.group(3)
            
            record = {'Date': date_str, 'District': district, 'Qty': end-start+1}
            records_part.append(record)
            
        res = newasym_pat1.match(line)
        if res:
            district = res.group(2)
            
            record = {'Date': date_str, 'District': district, 'Qty': 1}
            records_part.append(record)
        
        res = newasym_pat2.match(line)
        if res:
            start = int(res.group(1))
            end   = int(res.group(2))
            district = res.group(3)
            
            record = {'Date': date_str, 'District': district, 'Qty': end-start+1}
            records_part.append(record)
        
        if part_newpatn and "均为本市闭环隔离管控人员" in line:
            df_part = pd.DataFrame(records_part)
            records_part = []
            df_part['Cat'] = "NewPatn"
            df_part['From'] = "GL"
            
        if part_newpatn and "在风险人群筛查中发现新冠病毒核酸检测结果异常" in line:
            df_part = pd.DataFrame(records_part)
            records_part = []
            df_part['Cat'] = "NewPatn"
            df_part['From'] = "SX"
        
        if part_newpatn and "为此前报告的本土无症状感染者" in line:
            df_part = pd.DataFrame(records_part)
            records_part = []
            df_part['Cat'] = "NewPatn"
            df_part['From'] = "ZH"
            
        if part_newasym and "均为本市闭环隔离管控人员" in line:
            df_part = pd.DataFrame(records_part)
            records_part = []
            df_part['Cat'] = "NewAsym"
            df_part['From'] = "GL"
            
        if part_newasym and "在风险人群筛查中发现新冠病毒核酸检测结果异常" in line:
            df_part = pd.DataFrame(records_part)
            records_part = []
            df_part['Cat'] = "NewAsym"
            df_part['From'] = "SX"
        
        if not df_part.empty:
            if df.empty:
                df = df_part
            else:
                df = pd.concat([df, df_part])
            
            df_part = pd.DataFrame()

        res = newlvyy_pat.match(line)
        if part_newpatn and res:
            qty = int(res.group(1))
            records_newlvyy.append({"Date": date_str, "Sum": qty})
        
        res = newlvfc_pat1.match(line)
        if res:
            qty = int(res.group(1))
            records_newlvfc.append({"Date": date_str, "Sum": qty})
        
        res = newlvfc_pat2.match(line)
        if res:
            qty = int(res.group(1))
            records_newlvfc.append({"Date": date_str, "Sum": qty})
        
df['Date'] = df['Date'].apply(pd.Timestamp)
df = df.sort_values(by=["Date"])

df_newpos_caselst = df


df_newlvyy_casenum = pd.DataFrame(records_newlvyy)
df_newlvyy_casenum['Date'] = df_newlvyy_casenum['Date'].apply(pd.Timestamp)

df_newlvfc_casenum = pd.DataFrame(records_newlvfc)
df_newlvfc_casenum['Date'] = df_newlvfc_casenum['Date'].apply(pd.Timestamp)
```

### Sum by District

```python
df_convpatn_dist_casesum = df_newpos_caselst.loc[(df_newpos_caselst['Cat'] == "NewPatn") & (df_newpos_caselst['From'] == "ZH"), :].groupby(['Date', "District"]).sum().reset_index().rename(columns={"Qty": "ConvPatnSum"}).fillna(0)

# df_convpatn_dist_casesum
```

```python
df_newpatn_dist_casesum = df_newpos_caselst.loc[(df_newpos_caselst['Cat'] == "NewPatn") & (df_newpos_caselst['From'] != "ZH"), :].groupby(['Date', "District"]).sum().reset_index().rename(columns={"Qty": "NewPatnSum"}).fillna(0)

# df_newpatn_dist_casesum
```

```python
df_newasym_dist_casesum = df_newpos_caselst.loc[(df_newpos_caselst['Cat'] == "NewAsym"), :].groupby(['Date', "District"]).sum().reset_index().rename(columns={"Qty": "NewAsymSum"}).fillna(0)

# df_newasym_dist_casesum
```

```python
df_date_dist = pd.MultiIndex.from_product([df_newpos_caselst['Date'].unique(), df_dist_info['District'].unique()], names=["Date", "District"]).to_frame(index=False)

# df_date_dist
```

```python

df_newpos_dist_casesum = df_convpatn_dist_casesum.merge(df_newpatn_dist_casesum, on=["Date", "District"], how="outer").fillna(0)
df_newpos_dist_casesum = df_newpos_dist_casesum.merge(df_newasym_dist_casesum, on=["Date", "District"], how="outer").fillna(0)

df_newpos_dist_casesum['NewPosSum'] = df_newpos_dist_casesum['NewPatnSum'] + df_newpos_dist_casesum['NewAsymSum']

df_newpos_dist_casesum = df_date_dist.merge(df_newpos_dist_casesum, on=["Date", "District"], how="left").fillna(0)

# df_newpos_dist_casesum
```

```python
df_newpos_dist_casesum.to_excel("_date/_date_Dist_NewPos_CaseSum.xlsx", index=False)
```

### Sum by City

```python
df_convpatn_city_casesum = df_convpatn_dist_casesum.groupby(['Date']).sum().reset_index()

# df_convpatn_city_casesum
```

```python
df_newpatn_city_casesum = df_newpatn_dist_casesum.groupby(['Date']).sum().reset_index()

# df_newpatn_city_casesum
```

```python
df_newasym_city_casesum = df_newasym_dist_casesum.groupby(['Date']).sum().reset_index()

# df_newasym_city_casesum
```

```python
df_newpos_city_casesum = df_convpatn_city_casesum.merge(df_newpatn_city_casesum, on=["Date"], how="outer").fillna(0)
df_newpos_city_casesum = df_newpos_city_casesum.merge(df_newasym_city_casesum, on=["Date"], how="outer").fillna(0)

df_newpos_city_casesum['NewPosSum'] = df_newpos_city_casesum['NewPatnSum'] + df_newpos_city_casesum['NewAsymSum']

# df_newpos_city_casesum
```

```python
df_newlvs_city_casesum = df_newlvyy_casenum.merge(df_newlvfc_casenum, on="Date", how="outer").rename(columns={"Sum_x": "NewLvyySum", "Sum_y": "NewLvfcSum"}).fillna(0)

# df_newlvs_city_casesum
```

```python
df_city_casesum = df_newpos_city_casesum.merge(df_newlvs_city_casesum, on="Date", how="outer").sort_values(["Date"])

# df_city_casesum
```

```python
df_city_casesum.to_excel("_date/_date_City_CaseSum.xlsx", index=False)
```

## New Positive Locations

```python
newpos_loc_file_lst = glob.glob("LocationPR/*.txt")
```

```python
records = []
    
for filename in newpos_loc_file_lst:
    with open(filename) as f:
        content_raw = f.read()

    lines = content_raw.split("\n")
    
    date_str = "2022" + os.path.splitext(filename)[0].split('/')[-1]

    valid = False
    valid_lines = []

    district = None
    location = None

    location_lst_start = False
    location_valid = False    

    for line in lines:
        if line == '各区信息如下：':
            valid = True
        elif line == '来源：上海发布':
            valid = False

        if valid:
            if len(line) == 0 or line == ' ' or ('滑动' in line):
                pass
            else:
                valid_line = line.strip(',，：。、 ')

                location_lst_start = False

                if '居住于' in valid_line:
                    district = valid_lines[-1]
                    location_lst_start = True
                elif '消毒' in valid_line:
                    location_valid = False

                if location_valid:
                    
                    locations = valid_line.split("、")
                    
                    for loc in locations:
                        loc = loc.strip(' ')
                        if len(loc) > 0:
                            row = {'Date': date_str, 'District': district, 'Location': loc}
                            records.append(row)

                if location_lst_start:
                    location_valid = True

                valid_lines.append(valid_line)
            

df = pd.DataFrame(records)

df['Date'] = df['Date'].apply(pd.Timestamp)
df['NewPositive'] = True
df = df.sort_values(by=["Date"])

df_newpos_loclst = df
```

```python
df_newpos_locqty = df_newpos_loclst.groupby(['Date', 'District',]).count().reset_index().rename(columns={"Location": "LocQty"})
```

## Ever Positive Locations

```python
# Fengkong

df = pd.DataFrame()

for i in range(FENGKONG_START_DAY, FENGKONG_END_DAY + 1):
    df_tmp = df_newpos_loclst.copy().drop(columns=["NewPositive"])
    df_tmp['Date'] = df_newpos_loclst['Date'] + pd.offsets.Day(i)
    df_tmp['FlagFengkong'] = True

    if df.empty:
        df = df_tmp
    else:
        df = pd.concat([df, df_tmp])

df = df.loc[df['Date'] < pd.Timestamp.today().normalize(), :]
df = df.sort_values(["Date"]).drop_duplicates(subset=['Date', 'District', 'Location'], keep="first").reset_index(drop=True)

df_fk_loclst = df
```

```python
# Guankong

df = pd.DataFrame()

for i in range(GUANKONG_START_DAY, GUANKONG_END_DAY + 1):
    df_tmp = df_newpos_loclst.copy().drop(columns=["NewPositive"])
    df_tmp['Date'] = df_newpos_loclst['Date'] + pd.offsets.Day(i)
    df_tmp['FlagGuankong'] = True

    if df.empty:
        df = df_tmp
    else:
        df = pd.concat([df, df_tmp])

df = df.loc[df['Date'] < pd.Timestamp.today().normalize(), :]
df = df.sort_values(["Date"]).drop_duplicates(subset=['Date', 'District', 'Location'], keep="first").reset_index(drop=True)

df_gk_loclst = df
```

```python

df = pd.DataFrame()

df = df_fk_loclst.merge(df_gk_loclst, on=["Date", "District", "Location"], how="outer")

df['PreTestStatus'] = np.nan
df.loc[df['FlagFengkong'] == True, 'PreTestStatus'] = "FK"
df.loc[pd.isna(df['FlagFengkong']) & (df['FlagGuankong'] == True), 'PreTestStatus'] = "GK"

df = df.drop(columns=["FlagFengkong", "FlagGuankong"])

df_fkgk_loclst = df
```

```python
df = pd.DataFrame()

df = df_fkgk_loclst.merge(df_newpos_loclst, on=["Date", "District", "Location"], how="outer")

df['PostTestStatus'] = df['PreTestStatus']
df.loc[df['NewPositive'] == True, 'PostTestStatus'] = "FK"

df = df.sort_values("Date")

df_everpos_loclst = df
```

```python
df_fk_locqty = df_everpos_loclst.loc[df_everpos_loclst['PostTestStatus'] == "FK", :].groupby(['Date', 'District',]).count().reset_index().rename(columns={"Location": "LocQty"})
```

```python
df = df_everpos_loclst.loc[(df_everpos_loclst['PreTestStatus'] != "FK") & (df_everpos_loclst['NewPositive'] == True), :]
df = df.sort_values("Date").groupby(['Date', 'District',]).count().reset_index().rename(columns={"Location": "LocQty"})

df_newpos_ofk_locqty = df
```

```python
df = df_everpos_loclst.loc[pd.isna(df_everpos_loclst['PreTestStatus']) & (df_everpos_loclst['NewPositive'] == True), :]
df = df.sort_values("Date").groupby(['Date', 'District',]).count().reset_index().rename(columns={"Location": "LocQty"})

df_newpos_ofkogk_locqty = df
```
